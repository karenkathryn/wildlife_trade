## EDA

Count how many entries are for the year 2016. Compute a ratio against the total number of entries.

```python
df_2016 = df[df['Year'] == 2016]
df_2016.sample()
```

```python
print('The number of entries for 2016 is: ', len(df_2016))
```

```python
len(df_2016)/len(df)
```

So, most entries are from 2016. Let's just focus on those records. 

### Questions
Are there major importers or exporters?


```python
df_2016['Importer'].value_counts()
```

The US was the top importer in 2016. Let's return the top three.

```python
df_2016['Importer'].value_counts().nlargest(3)
```

Now look at exporters.

```python
df_2016['Exporter'].value_counts()
```

NL was the major exporter. From the data dictionary, this is the Netherlands.

The reported purpose of the transaction is shown as a one-letter code: I wonder for what porpose did the US import?

```python
df_us_purpose = df_2016[df_2016['Importer']=='US'].groupby(['Purpose']).count()
df_us_purpose = df_us_purpose[[i for i in list(df_us_purpose.columns) if i == 'Importer reported quantity']]
df_us_purpose
```

Look at Japan.

```python
df_jp_purpose = df_2016[df_2016['Importer']=='JP'].groupby(['Purpose']).count()
df_jp_purpose = df_jp_purpose[[i for i in list(df_us_purpose.columns) if i == 'Importer reported quantity']]
df_jp_purpose
```

Now Germany

```python
df_de_purpose = df_2016[df_2016['Importer']=='DE'].groupby(['Purpose']).count()
df_de_purpose = df_de_purpose[[i for i in list(df_us_purpose.columns) if i == 'Importer reported quantity']]
df_de_purpose
```

Hm, that a lot of info. Let's narrow this down, more. According to the data dictionary, 'Z' represents Zoo.

Find the top three countries that reported zoo imports.



```python
x = df_2016[(df_2016['Purpose']=='Z')].groupby(['Importer']).count()
x = x.drop('Year',axis=1)
x = x.drop('Exporter reported quantity',axis=1)
y = x['Importer reported quantity'].nlargest(3)
y
```

That's France, China, and US. Let's break these numbers down by Term.

```python
#France 
df_2016[(df_2016['Importer']=='FR') & 
        (df_2016['Purpose']=='Z')].groupby(['Term']).count()
```

```python
#China
df_2016[(df_2016['Importer']=='CN') & 
        (df_2016['Purpose']=='Z')].groupby(['Term']).count()
```

```python
#US
df_2016[(df_2016['Importer']=='US') & 
        (df_2016['Purpose']=='Z')].groupby(['Term']).count()
```

Let's make some vizualizations out of this information.
* show the comparison between the total numbers

```python
# rot == Rotation for ticks 
y.plot.bar(x='id', y='Importer reported quantity', rot=0, title='Top 3 importers for category Zoo');
```

Does a pie chart look better? Not really.

```python
ax = y.plot(x='id', y='Importer reported quantity', kind='pie');
```
